```
spark=SparkSession.builder.master('local[1]').getOrCreate()
spark.conf.set('spark.sql.repl.eagerEval.enabled',True)
```

```
sc=SparkContext.getOrCreate(SparkConf().setMaster('local[1]'))
```

```
def isprime(n):
    n=abs(n)
    if n<2: 
        return False
    elif(not n & 1):
        return False
    elif (n==2):
        return True
    for x in range(3,int(n**0.5)+1,2):
        if n%x==0:
            return False
        else:
            return True
```

```
num=sc.parallelize(range(10000))
print(num.filter(isprime).count())
```

```
text=sc.textFile("/FileStore/tables/bda.txt")
text.collect()
```

```
from operator import add
def token(text):
    return text.split()
```

```
words=text.flatMap(token)
words.collect()
```

```
wc=words.map(lambda x: (x,1))
wc.collect()
```

```
wc.toDebugString()
```

```
counts = wc.reduceByKey(add)
counts.collect()
```

```
counts.saveAsTextFile("wcc1")
```

```
counts.collect()
```
